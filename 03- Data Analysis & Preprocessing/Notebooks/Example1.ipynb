```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = load_iris()
df = pd.DataFrame(data['data'], columns=['a','b','c','d'])
```

```python
df.hist()
plt.tight_layout()
plt.show()
```

```python
df.corr().style.background_gradient(cmap='coolwarm')
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].scatter(df['a'],df['c'])
ax[0,1].scatter(df['a'],df['d'])
ax[1,0].scatter(df['c'],df['d'])
ax[1,1].scatter(df['b'],df['c'])
plt.tight_layout()
plt.show()
```

```python
s = pd.Series(data['target'])
plt.bar(s.unique(), s.value_counts())
```
